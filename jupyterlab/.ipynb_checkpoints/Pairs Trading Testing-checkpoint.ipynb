## Implement Statistics Package

```python
import numpy as np
import pandas as pd

import statsmodels
import statsmodels.api as sm
import statsmodels.tsa.stattools.coint as coint

import  matplotlib.pyplot as plt
```

## Generate Fake Securities
* Cointegration of stationary series is easy here, but markets don't reflect this all the time
* Important to defintine stationary and non-stationary series.

```python
## Generate Fake Securities

np.random.seed(107) # generate seed for randomness
x_returns = np.random.normal(0, 1, 100) #100 points, deviates by 1
x = pd.Series(np.cumsum(x_returns), name='x') + 50
x.plot();

some_noise = np.random.normal(0, 1, 100)
y = x + 5 + some_noise
y.name = 'y'
pd.concat([x, y], axis=1).plot();
```

## Cointegrate Securities

* More subtle relationship than correlation
* Finds linear combination around mean
* Finding probability distribution

```python
# Cointegrate
(y - x).plot() # Plot Spread
plt.axhline((y - x).mean(), color = 'black', linestyle='-') # mean spread/distance
plt.xlabel('time')
plt.ylabel('spread')
plt.legend(['price spread', 'mean']);
```

## Test Cointegration
* Compute p-value to test 'stationary' between time-series around mean

```python
score, pvalue, _ = coint(x,y)
print pvalue
```
